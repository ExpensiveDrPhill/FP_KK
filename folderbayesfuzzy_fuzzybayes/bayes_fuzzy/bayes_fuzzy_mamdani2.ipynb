```python
import pandas as pd
from importlib import util
import sys
import bayes_module as bayes  # jika modul ini menyimpan risk_overload()

# load fuzzy mamdani
spec = util.spec_from_file_location("fuzzy_mamdani_module", "/content/fuzzy_mamdani (1).py")
fuzzy_mod = util.module_from_spec(spec)
sys.modules["fuzzy_mamdani_module"] = fuzzy_mod
spec.loader.exec_module(fuzzy_mod)
mamdani_decision = fuzzy_mod.mamdani_decision

# load dataset
df = pd.read_csv("PowerLoad_Dataset.csv")
df['Hour'] = pd.to_datetime(df['Timestamp']).dt.hour

# test 10 rows
for i in range(10):
    row = df.iloc[i]

    load = row['Power_Load_kW']
    proj = row['Weekly_PreDispatch_Projection']
    prob = bayes.risk_overload(load, proj, scale=50)

    dev = row['Power_Load_kW'] - row['Weekly_PreDispatch_Projection']
    adj = mamdani_decision(prob, dev)


    print(f"Load={load:.1f} Pred={proj:.1f} Dev={load-proj:.1f} "
          f"=> Prob={prob:.3f}, Adjustment={adj:.3f}")
```
