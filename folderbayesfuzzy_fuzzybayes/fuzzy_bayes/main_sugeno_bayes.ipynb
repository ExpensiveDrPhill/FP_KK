```python
# main.py
import pandas as pd
from bayes_module import bayes_prob_overload
from fuzzy_sugeno import sugeno_first_order_scalar, apply_to_dataframe

df = pd.read_csv("PowerLoad_Dataset.csv")
df['Hour'] = pd.to_datetime(df['Timestamp']).dt.hour

# compute probs for all rows (vectorization depends on bayes_module)
probs = []
for idx, row in df.iterrows():
    probs.append(bayes_prob_overload(row['Hour'], row['DayOfWeek'], row['Precipitation_mm']))
df['Prob'] = probs

# gunakan apply_to_dataframe dengan koefisien default, bisa diganti
df = apply_to_dataframe(df, prob_col='Prob', out_col='Sugeno',
                        p1=-0.5, q1=0.4, p2=0.7, q2=0.2)

# tampilkan 10 teratas
print(df[['Hour','DayOfWeek','Precipitation_mm','Prob','Sugeno']].head(10))

# simpan hasil
df.to_csv("PowerLoad_with_Sugeno.csv", index=False)
print("Saved to PowerLoad_with_Sugeno.csv")

```
