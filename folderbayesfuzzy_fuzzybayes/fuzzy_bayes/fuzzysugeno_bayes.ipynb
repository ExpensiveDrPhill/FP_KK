```python
import pandas as pd
from sugeno_module import GridEarlyWarning_Sugeno
from bayesmodule import GridBayesEngine

df = pd.read_csv("PowerLoad_Dataset.csv")

# Preprocess
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Hour'] = df['Timestamp'].dt.hour

# Load engines
sugeno = GridEarlyWarning_Sugeno()
bayes = GridBayesEngine()

print(f"{'TIMESTAMP':<20} | {'SUGENO%':<10} | STATUS")

rows_to_process = min(40, len(df))

for i in range(rows_to_process):
    row = df.iloc[i]

    # Inputs
    t_val = float(row['Hour'])
    d_val = float(row['DayOfWeek']) - 1.0
    r_val = float(row['Precipitation_mm'])

    # Fuzzy Sugeno score
    score_s = sugeno.evaluate(t_val, d_val, r_val)

    # Bayes probability
    prop_m = bayes.get_failure_probability(score_s)

    # Status
    if prop_m >= 0.8:
        status = "CRITICAL"
    elif prop_m >= 0.3:
        status = "WARNING"
    else:
        status = "NORMAL"

    print(f"{str(row['Timestamp']):<20} | {prop_m:.1%}    | {status}")
```
